## **Relationships Between Tables in CreditScoreDB**  

The **CreditScoreDB** database is structured to manage customer financial data, loan history, and credit evaluations efficiently. Below is a detailed explanation of the relationships between the tables:

### **1ï¸âƒ£ Customers â†’ Loan_Applications (1:M)**

- **Foreign Key:** `Loan_Applications.customer_id â†’ Customers.customer_id`

### **2ï¸âƒ£ Customers â†’ Loan_History (1:M)**
- **Relationship:** A **customer** can have multiple **approved loans**, but each loan is linked to only **one customer**.  
- **Foreign Key:** `Loan_History.customer_id â†’ Customers.customer_id`  

ğŸ“Œ **Example:** A customer with **an active home loan** and a **closed car loan**.  

---

### **3ï¸âƒ£ Customers â†’ Credit_Scores (1:M)**
- **Relationship:** A **customer** can have multiple **credit score evaluations** over time, but each credit score record belongs to **one customer**.

ğŸ“Œ **Example:** A customerâ€™s **credit score can change** based on loan repayments.

### **4ï¸âƒ£ Customers â†’ Credit_Inquiries (1:M)**

ğŸ“Œ **Example:** A bank checking a customerâ€™s credit history before approving a **new loan**.

### **5ï¸âƒ£ Loan_History â†’ Payment_History (1:M)**
- **Relationship:** A **loan** can have multiple **payments**, but each payment belongs to **only one loan**.  
- **Foreign Key:** `Payment_History.loan_id â†’ Loan_History.loan_id`

### **6ï¸âƒ£ Loan_History â†’ Collateral (1:1)**

ğŸ“Œ **Example:** A **car loan** backed by the **car itself** as collateral.

### **7ï¸âƒ£ Payment_History â†’ Late_Payment_Penalty (1:M)**
- **Relationship:** A **payment** can have multiple **late payment penalties**, but each penalty is associated with **only one payment**.

### **8ï¸âƒ£ Loan_Applications â†’ Loan_History (1:1)**
- **Relationship:** A **loan application** is either **approved** and becomes a **loan**, or **rejected**. If approved, it corresponds to **one loan record** in Loan_History.  
- **Foreign Key:** `Loan_History.customer_id â†’ Loan_Applications.customer_id`  

ğŸ“Œ **Example:** A **car loan application** that was **approved** and added to `Loan_History`.

## ** Database Normalization Testing**

### **ğŸ”¹ 1NF (First Normal Form) Testing**
#### **ğŸ“ Problem Statement:**
Verify that all tables in `CreditScoreDB` follow **1NF rules**, ensuring:
1. Each column contains **atomic (indivisible) values**.
2. Each column contains values of **a single data type**.
3. Each row has a **unique identifier (Primary Key)**.

#### **ğŸ“ Testing Approach:**
- Verified that no tables contain **repeating groups** or **multi-valued attributes**.
- Checked if every column has **atomic values** (e.g., `first_name` and `last_name` are separate fields).

#### **âœ… Result:**
âœ” The dataset **passes 1NF** since:
- All tables have a **Primary Key**.
- No **multi-valued attributes** exist.
- All values are **atomic and properly structured**.

---

### **ğŸ”¹ 2NF (Second Normal Form) Testing**
#### **ğŸ“ Problem Statement:**
Ensure that all tables in `CreditScoreDB` comply with **2NF**, requiring:
1. The database is already in **1NF**.
2. All **non-key attributes** are **fully dependent** on the **Primary Key** (no partial dependency).

#### **ğŸ“ Testing Approach:**
- Checked if tables with **composite primary keys** have attributes that depend on **only part of the key**.
- Ensured all **non-key attributes** are directly related to the **whole primary key**, not just a part of it.

#### **âœ… Result:**
âœ” The dataset **passes 2NF** since:
- No **partial dependencies** exist in any table.
- All non-key attributes are fully functionally dependent on the **Primary Key**.

---

### **ğŸ”¹ 3NF (Third Normal Form) Testing**
#### **ğŸ“ Problem Statement:**
Verify that all tables meet **3NF**, ensuring:
1. The database is already in **2NF**.
2. There are **no transitive dependencies** (a non-key attribute should not depend on another non-key attribute).

#### **ğŸ“ Testing Approach:**
- Checked if any **non-key attributes** depend on another **non-key attribute** instead of the **Primary Key**.
- Verified that attributes are **directly dependent** on the **Primary Key** and **not on other fields**.

#### **âœ… Result:**
âœ” The dataset **passes 3NF** since:
- There are **no transitive dependencies**.
- All non-key attributes are **directly dependent on the Primary Key**.

#### **ğŸ“ Interpretation:**
âœ” The database structure is optimized to prevent anomalies, ensuring **better efficiency, faster queries, and easier maintenance**.
